- Normalización
- Valores faltantes

```python
# Normalización
df_scaled = pd.DataFrame(preprocessing.scale(df))
df_scaled.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Incluir valores faltantes
# Primero creamos unos datos de ejemplo
missing = [[1, 2], [np.nan, 3], [7, 6]]
missing
```

```python
# Creamos el objeto
replace_missing_with_mean = Imputer(missing_values='NaN', strategy='mean', axis=0)
# Aplicamos a los datos
replace_missing_with_mean.fit(missing)
replace_missing_with_mean.transform(missing)
```

```python
# Para tranformar datos categoricos la función get_dummies de Pandas funciona muy bien
# Primero creamos uns datos de ejemplo
flights = {'flight' : pd.Series(['Barcelona','Madrid','Barcelona','Madrid'], index=[1, 2, 3, 4]),
               'price' : pd.Series([12., 21., 11., 23.], index=[1, 2, 3, 4])}
df_category=pd.DataFrame(flights)
df_category
```

```python
# Luego creamos un dataframe con las columnas con los atributos One-Hot encoded
tag=pd.get_dummies(df_category["flight"],prefix="flight")
# y unimos al dataframe inicial
df_category=pd.concat([df_category,tag], axis=1)
df_category
```

# Muestra de entrenamiento, validación y pruebas

- Entrenamiento tipicamente 80% de los datos
  - De esos 80% se pueden separar 20% del total de datos como validación
  - O utilizar validación k-fold - ejemplo se dividen los 80% en 4 partes iguales (4-fold) de 20% y se entrenan los en 3 de estas y se prueba en la 4. Eso se hace 4 veces con cada variación y se saca la media
- Pruebas los 20% restantes
  - Las pruebas solo se utilizan una vez que esté terminado el modelo. *Optimizar los modelos con el feedback de la muestra de pruebas es trampa*
